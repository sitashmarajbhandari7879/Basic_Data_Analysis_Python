```python
import pandas as pd
data = pd.ExcelFile('yelp.xlsx')
df = data.parse('yelp_data')

```

```python
df.head()
```

```python
df_cities = data.parse('cities')
df_states = data.parse('states')
```

```python
# Merging the sheets of excel file into one dataframe
df = pd.merge(left = df, right= df_cities, how = 'inner', left_on = 'city_id', right_on = 'id') 
df = pd.merge(left = df, right= df_states, how = 'inner', left_on = 'state_id', right_on = 'id') 
```

```python
df.head()
```

```python
df.shape
```

```python
# quering the name, city and state of the resturant
attributes = ['name','city','state','category_0']
df[attributes].head(100)
```

```python
# removing duplicate collumn
df = df.T.drop_duplicates().T
df.head()
```

```python
# retrieving last business in the dataset
df[-1:]['name']
```

```python
# filtering Dataframe using Boolean indexing
# selecting business from the city Bellevue
Bellevue= df['city'] == 'Bellevue'
df[Bellevue]
```

```python
# retriving a specific value 
resturant = df ['name'] == 'Rusty Nail'
df[resturant]
```

```python
# retriving all bars from the dataframe
cat=df['category_0']== 'Bars'
df[cat]
```

```python
# bars and resturant in the city of las vegas
cat_0 = df["category_0"]. isin(["Bars","Restaurant"])
cat_1 = df["category_1"]. isin(["Bars","Restaurant"])
vegas = df["city"] == 'Las Vegas'
df[(cat_0|cat_1) & vegas]
```

```python
# retrive all dive bar with 4 star and recommend one random
import random
lv = df ['city'] == 'Las Vegas'
cat_0_dbar = df['category_0']== 'Dive Bars'

cat_1_dbar = df['category_1']== 'Dive Bars'
divebar_lv = df [(cat_0_dbar|cat_1_dbar)& lv]
stars= divebar_lv['stars'] >= 4.0
divebar_lv_4 = divebar_lv[stars]
# to select random dive bar
rand = random.randint(0,len(divebar_lv_4)-1)
rand_divebar = divebar_lv_4[rand : rand + 1]
rand_divebar
```

```python
# another method
random_divebar = divebar_lv_4.iloc[rand]
random_divebar
```

```python
# adding new column
df['rating'] = df ['stars'] * 2
def convert_to_rating(x):
    return(str(x) + ' out of 10')
df['rating'] = df['rating'].apply(convert_to_rating)
df.head()
```
